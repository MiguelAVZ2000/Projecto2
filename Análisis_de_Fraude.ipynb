## **Cargar el Dataset**

```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv("creditcard.csv")

# Mostrar las primeras filas
print(df.head())

# Ver información general
print(df.info())
```

## **Exploración y Limpieza de Datos**

```python
# Verificar valores nulos
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contar la cantidad de transacciones fraudulentas y no fraudulentas
plt.figure(figsize=(6,4))
sns.countplot(x=df["Class"], palette="coolwarm")
plt.title("Distribución de Transacciones (0 = No Fraude, 1 = Fraude)")
plt.show()

# Ver proporciones
fraud_ratio = df["Class"].value_counts(normalize=True) * 100
print(f"Proporción de transacciones fraudulentas: {fraud_ratio[1]:.4f}%")
```

```python
print(df.describe())
```

## **Análisis Exploratorio (EDA)**

```python
import numpy as np

# Graficar la distribución de algunas variables
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, feature in enumerate(["V1", "V2", "V3"]):  # Se pueden cambiar por otras variables
    sns.histplot(df[feature], bins=50, kde=True, ax=axes[i])
    axes[i].set_title(f"Distribución de {feature}")

plt.show()
```

```python
fraud = df[df["Class"] == 1]
no_fraud = df[df["Class"] == 0]

# Comparar estadísticos de algunas variables
features = ["V1", "V2", "V3", "Amount"]
print("Media de transacciones fraudulentas:")
print(fraud[features].mean())

print("\nMedia de transacciones normales:")
print(no_fraud[features].mean())
```

## **Visualizar Patrones en los Fraudes**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Class"], y=df["Amount"], palette="coolwarm")
plt.yscale("log")  # Escala logarítmica para mejorar la visualización
plt.title("Distribución del Monto en Transacciones (0 = No Fraude, 1 = Fraude)")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Mapa de Calor de la Correlación entre Variables")
plt.show()
```

## **Preparación de Datos para el Modelo**

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE

# Evitar advertencias de joblib
os.environ["LOKY_MAX_CPU_COUNT"] = "4"

X = df.drop(columns=["Class"])
y = df["Class"]

# Aplicar SMOTE
smote = SMOTE(sampling_strategy=0.2, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Ver nueva distribución
sns.countplot(x=y_resampled, palette="coolwarm")
plt.title("Distribución después del Sobremuestreo")
plt.show()
```
